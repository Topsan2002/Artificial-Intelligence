```python
import pandas as pd
path = '/content/drive/MyDrive/Colab Notebooks/DataSet/people-2000000.csv'
dfp = pd.read_csv(path) ## Dataset is now stored in a Pandas Dataframe
dfp
```

```python
dfp.to_parquet('people_data.parquet')
table = pa.Table.from_pandas(dfp)
table.drop(['User Id', 'Email','Date of birth' ])
```

```python
pq.write_to_dataset(table, root_path='my_parq', partition_cols=['Job Title'])
```

ข้อที่ 1

```python
table = pq.read_table('my_parq',columns=['Index', 'First Name','Last Name','Sex','Job Title'])
table.to_pandas()
```

```python
table
```

ข้อที่ 2

```python
import pyarrow.compute as pc
grouped_table = table.group_by("Job Title").aggregate(
  [("Index",
  "count",
  pc.CountOptions(mode="only_valid"))]
)
grouped_table.to_pandas()
```

ข้อที่ 3

```python
expr1 = pc.field("Sex") == "Male"
expr2 = pc.field("Job Title") == "Writer"
table.select(['Last Name', 'Sex', 'Job Title']).filter(expr1 & expr2).to_pandas()
```
